```python
def analyze_post(likes):
    total_likes = 0

    # Loop to calculate total likes
    for like in likes:
        total_likes += like

    # Print in required format
    print(f"Total Likes: {total_likes}")

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# User input
n = int(input())
likes = []

for i in range(n):
    likes.append(int(input()))

analyze_post(likes)
```

```python
# Function to monitor stock
def check_stock(stock):
    print(f"Medicine Stock: {stock}")

    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


# Take input from user
stock = int(input())

# Call function
check_stock(stock)
```

```python
# Function to analyze rainfall
def analyze_rainfall(data, required):
    total = 0

    # Loop to calculate total rainfall
    for value in data:
        total += value

    # Calculate average
    average = total // len(data)

    print(f"Average Rainfall: {average}")

    if average >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Take input
n = int(input())  # number of rainfall entries
rainfall = []

for i in range(n):
    rainfall.append(int(input()))

required_level = int(input())  # required rainfall level

# Call function
analyze_rainfall(rainfall, required_level)
```

```python
# Function to check duplicates
def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Take input
n = int(input())
usernames = []

for i in range(n):
    usernames.append(input())

# Call function
check_duplicates(usernames)
```

```python
# Function to check eligibility
def check_eligibility(age):
    print(f"Patient Age: {age}")

    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")


# Take input from user
age = int(input())

# Call function
check_eligibility(age)
```

```python
# Function to filter premium crops
def filter_premium(prices):
    premium = []

    # Loop to filter prices
    for price in prices:
        if price > 2000:
            premium.append(price)

    print(f"Premium Crops: {premium}")


# Take input
n = int(input())
prices = []

for i in range(n):
    prices.append(int(input()))

# Call function
filter_premium(prices)
```

```python
# Function to evaluate system health
def evaluate_health(errors):
    print(f"Error Count: {errors}")

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Take input from user
errors = int(input())

# Call function
evaluate_health(errors)
```

```python
# Function to validate transaction
def validate_transaction(amount):
    print(f"Transaction Amount: {amount}")

    if amount <= 50000:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Take input from user
amount = int(input())

# Call function
validate_transaction(amount)
```

```python
# Function to calculate attendance and check eligibility
def check_attendance(records):
    total_days = len(records)
    present_days = 0

    # Loop to count present days
    for day in records:
        if day == 1:
            present_days += 1

    # Calculate percentage
    percentage = (present_days / total_days) * 100

    print(f"Attendance Percentage: {percentage}")

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Input
n = int(input())  # total days
attendance = []

for i in range(n):
    attendance.append(int(input()))  # 1 = present, 0 = absent

# Call function
check_attendance(attendance)
```

```python
# Function to calculate bill and usage status
def calculate_bill(units):
    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Input from user
units = int(input())

# Function call
total_bill, usage_status = calculate_bill(units)

# Output
print(f"Total Bill: {total_bill}")
print(f"Usage Level: {usage_status}")
```
