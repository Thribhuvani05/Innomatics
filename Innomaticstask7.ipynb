```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display(self):
        print(f"{self.name} is {self.status}")

light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ").upper()

light = SmartLight(light_name)

if action == "ON":
    light.turn_on()
else:
    light.turn_off()

light.display()
```

```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)

name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
dept = input("Enter Department: ")

emp = Employee(name, emp_id, dept)
emp.display()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

contact = Contact(name, phone)
contact.display()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print(f"Price: ₹{self.price}")

name = input("Enter Product Name: ")
price = input("Enter Price: ")

product = Product(name, price)
product.display()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print(f"Rating: {self.rating} / 5")


# User Input
name = input("Enter Movie Name: ")
rating = input("Enter Rating: ")

movie = Movie(name, rating)
movie.display()
```

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)

customer = input("Enter Customer Name: ")
address = input("Enter Address: ")

delivery = Delivery(customer, address)
delivery.display()
```
