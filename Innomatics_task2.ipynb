```python
#Given a sentence, identify all unique words and display:

sentence = input("Enter Sentence:")
words = sentence.split()
unique_words = set(words)
print("unique words count",len(unique_words))
print("unique words",unique_words)
```

```python
#Store employee names and their salaries, and determine which employee has the highest salary.

n = int(input("Enter number of employees: "))
employees = {}
for i in range(n):
  name = input("Enter employee name: ")
  salary = int(input("Enter employee salary: "))
  employees[name] = salary
  highest_employee = max(employees,key = employees.get)
highest_salary = max(employees.values())
print("highest salary: ",highest_employee,"-",highest_salary)
```

```python
#Find the maximum and minimum values from a list of numbers.

numbers = list(map(int,input("Enter numbers: ").split()))
maximum = max(numbers)
minimum = min(numbers)
print("List:",numbers)
print("Maximum:",maximum)
print("Minimum:",minimum)

```

```python
#count how many product prices are greater than 1000
prices = list(map(int,input("Enter prices: ").split()))
count =0
for price in prices:
  if price>1000:
    count+=1
print("Products above 1000: ",count)
```

```python
#calculate attendance percentage
attendance = input("Enter attendance (P/A seperated by space): ").split()
total_days = len(attendance)
present_count = attendance.count("P")
percentage = (present_count / total_days)*100
print("Attendance Perentage:",percentage)
```

```python
#remove duplicate phone numbers and display unique ones
phone_numbers = list(map(int,input("Enter phone numbers:").split()))
unique_numbers = set(phone_numbers)
print("Unique phone numbers: ",unique_numbers)
```

```python
#count how many times each character appears in a string
text = input("Enter a string: ")

char_count = {}

for ch in text:
    char_count[ch] = char_count.get(ch, 0) + 1
print(char_count)
```
